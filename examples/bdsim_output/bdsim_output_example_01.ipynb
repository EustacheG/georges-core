```python

def discover_histograms(histograms):
    return {h.name.decode('utf-8'): i for i, h in enumerate(histograms[0])}

def merge_histograms(histograms, index):
    _ = np.array([
        h[index].values for h in histograms
    ])
    return np.mean(_, 1), np.std(_, 1)
```
